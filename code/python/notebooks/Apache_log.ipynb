### 匹配日志，创建字典列表

```python
import apachelog, sys
```

设置格式

```python
fformat = r'%V %h %l %u %t \"%r\" %>s %b \"%i\" \"%{User-Agent}i\" %T'
```

```python
p = apachelog.parser(fformat)
```

```python
log = open('access_log').readlines()
```

匹配每一行，创建一个字典的列表

```python
log_list = []
for line in log:
       try:
          data = p.parse(line)
       except:
          sys.stderr.write("Unable to parse %s" % line)
       #时间格式稍微转化下，便于 pandas 去匹配
       data['%t'] = data['%t'][1:12]+' '+data['%t'][13:21]
       log_list.append(data)
```

### 创建并调整Data Frame

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame, Panel
```

```python
df = DataFrame(log_list)
```

删除一些暂时不需要的项

```python
del df['%T']; del df['%V']; del df['%i']; del df['%l']; del df['%u']; del df['%{User-Agent}i']
```

重命名 Data Frame 相关列

```python
df = df.rename(columns={'%>s': 'Status', '%b':'b', 
                        '%h':'IP', '%r':'Request', '%t': 'Time'})
```

显示前5行

```python
df.head()
```

将 Time 这一栏转换为 datatime 格式，并作为索引

```python
df.index = pd.to_datetime(df.pop('Time'))
```

```python
df['Status'] = df['Status'].astype('int')
```

将‘-’转化为 NaN，同时可根据需要将 bytes 换算为 Mb

```python
def dash2nan(x):
    if x == '-':
        x = np.nan
    else:
        x = float(x)/1048576.
    
    return x
```

```python
df['b'] = df['b'].apply(dash2nan)
```

### 流量分析
